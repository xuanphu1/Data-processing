## 3. HÀM VẼ ĐỒ THỊ HỒI QUY TUYẾN TÍNH

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
def Linear_regression(Data_type) :
    x = np.arange(0,24).reshape((-1,1))
    y = np.array(Data_type)
    model = LinearRegression().fit(x,y)
    r_sq = model.score(x,y)
    plt.xticks(np.arange(0,24,2))
    plt.text(25,y[len(y)-1],"y = " + str(model.coef_[0])+ "X" + " + " + str(model.intercept_) + "\n" + "R = " + str(r_sq))
    plt.plot(x,y,"go")
    plt.plot([min(x), max(x)],[model.coef_ * min(x) + model.intercept_, model.coef_ * max(x) + model.intercept_],'g--',color="red")
    plt.show()
```

## 4. HÀM TRẢ VỀ SỐ NGÀY TRONG TỪNG THÁNG

```python
def day_in_month(Month,year) :
    if Month == 1 or Month == 3 or Month == 5 or Month == 7 or Month == 8 or Month == 10 or Month == 12 :
            return 31
    elif Month == 4 or Month == 6 or Month == 9 or Month == 11 :
            return 30
    elif  year % 100  == 0 :
        if year % 400 == 0 :
            return 28
        else :
            return 29
    else :
        if year % 4 == 0 :
            return 29
        else :
            return 28
```

## 5. HÀM XỬ LÝ TRẢ VỀ MAX , MIN DATA TRONG TỪNG THÁNG VÀ NĂM

```python
def Max_Min_Data_in_the_month(Data_type, select ="") :
    Maximum_value_of_each_hour =["Max month :"]
    Minimum_value_of_each_hour =["Min month :"]
    for i in range(24) :
        Maximum_value_of_each_hour.append(0)
        Minimum_value_of_each_hour.append(1000)
    for i in range(1,day_in_month(i,year) + 1 ) :
        for j in range(1,25) :
            Maximum_value_of_each_hour[j] = max(Maximum_value_of_each_hour[j],Data_type[i][j])
            if Data_type[i][j] == 0 :
                continue
            else :
                Minimum_value_of_each_hour[j] = min(Minimum_value_of_each_hour[j],Data_type[i][j])
    if select == "Min" :
        return Minimum_value_of_each_hour[1:]
    elif select == "Max" :
        return Maximum_value_of_each_hour[1:]
#-------------------------------------------------------------------------------------------------------
def Max_Min_Data_in_the_year(Data_type,select="") :
    Maximum_value_of_each_hour = ["Max year :"]
    Minimum_value_of_each_hour = ["Min year :"]
    for i in range(24) :
        Maximum_value_of_each_hour.append(0)
        Minimum_value_of_each_hour.append(1000)
    for i in range(1,13) :
        for j in range(1,day_in_month(i,year) + 1 ) :
            for k in range(1,25) :
                Maximum_value_of_each_hour[k] = max(Maximum_value_of_each_hour[k],Data_type[i][j][k])
                if Data_type[i][j][k] == 0 :
                    continue
                else : 
                    Minimum_value_of_each_hour[k] = min(Minimum_value_of_each_hour[k],Data_type[i][j][k])
    if select == "Min" :
        return Minimum_value_of_each_hour[1:]
    elif select =="Max" :
        return Maximum_value_of_each_hour[1:]
```

## 6. HÀM TÌM MODE ( TẦN SUẤT )

```python
import statistics
# Hàm xử lý tìm mode trong cả năm -----------------------------------------------------
def Mode_data(Data_type_mode,choose_type ="") :
    Value_range_of_mode = [["Mode :"]]
    Frequency_of_the_value = ["Frequency :"]
    for i in range(24) :
        Value_range_of_mode.append(statistics.mode(Data_type_mode[i]))
        Frequency_of_the_value.append(Data_type_mode[i].count(statistics.mode(Data_type_mode[i])))
    if choose_type == "Frequency" :
        return Frequency_of_the_value[1:]
    elif choose_type == "Value" :
        return Value_range_of_mode[1:]
#-------------------------------------------------------------------------
```

```python
from os import path
def Data_processing(Incomplete_file_name ='C:\\Users\\DELL\\Downloads\\data_new\\may1\\2022-',year = 2022) :
    check_existence_of_data_for_the_month = False
    # ---------------------------------------------------------------------------------------
    Data_array_find_the_mode_of_temperature_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_humidity_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_pressure_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_PM1_0_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_PM2_5_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_PM10_0_in_the_month.append(["Year : "  +str(year)])
    Data_array_find_the_mode_of_CO2_PPM_in_the_month.append(["Year : "  +str(year)])

    # Lưu dữ liệu trung bình trong từng tháng 
    Average_Temperature_Data_for_month = []
    Average_Humidity_Data_for_month = []
    Average_Pressure_Data_for_month = []
    Average_PM1_0_Data_for_month = []
    Average_PM2_5_Data_for_month = []
    Average_PM10_0_Data_for_month = []
    Average_CO2_PPM_Data_for_month = []
    #Lưu tổng dữ liệu trong từng giờ trong 1 năm
    Total_Temperature_data = []
    Total_Humidity_data = []
    Total_Pressure_data = []
    Total_PM1_0_data = []
    Total_PM2_5_data = []
    Total_PM10_0_data = []
    Total_CO2_PPM_data = []

    #Lưu tổng số phần tử dữ liệu trong mỗi giờ
    Total_number_of_data_elements = []
    
    #Lưu tổng số phẩn tử dữ liệu cả năm 
    Total_number_of_data_elements_in_the_year = [["Year :" + str(year)]]
    
    
    for i in range(0,24) :
        Total_number_of_data_elements.append(0)
        Total_Temperature_data.append(0)
        Total_Humidity_data.append(0)
        Total_Pressure_data.append(0)
        Total_PM1_0_data.append(0)
        Total_PM2_5_data.append(0)
        Total_PM10_0_data.append(0)
        Total_CO2_PPM_data.append(0)
    # Month là vòng for chạy tháng
    for Month in range(1,13) :
        # Dữ liệu dùng để tìm mode
        Array_of_hourly_temperature_data = []
        Array_of_hourly_humidity_data = []
        Array_of_hourly_pressure_data = []
        Array_of_hourly_PM1_0_data = []
        Array_of_hourly_PM2_5_data = []
        Array_of_hourly_PM10_0_data = []
        Array_of_hourly_CO2_PPM_data = []
        for i in range(24) :
            Array_of_hourly_temperature_data.append(0)
            Array_of_hourly_humidity_data.append(0)
            Array_of_hourly_pressure_data.append(0)
            Array_of_hourly_PM1_0_data.append(0)
            Array_of_hourly_PM2_5_data.append(0)
            Array_of_hourly_PM10_0_data.append(0)
            Array_of_hourly_CO2_PPM_data.append(0)

            # Tạo mảng lưu dũ liệu theo giờ 

            Array_of_hourly_temperature_data[i] = []
            Array_of_hourly_humidity_data[i] = []
            Array_of_hourly_pressure_data[i] = []
            Array_of_hourly_PM1_0_data[i] = []
            Array_of_hourly_PM2_5_data[i] = []
            Array_of_hourly_PM10_0_data[i] = []
            Array_of_hourly_CO2_PPM_data[i] = []
        # -------------------------------------------------------------
        NULL = []
        Number_of_data_elements_in_the_month = [["Month : " + str(Month)]]
        Temperature_data_for_the_month = [["Month : " + str(Month)]]
        Humidity_data_for_the_month = [["Month : " + str(Month)]]
        Pressure_data_for_the_month = [["Month : " + str(Month)]]
        PM1_0_data_for_the_month = [["Month : " + str(Month)]]
        PM2_5_data_for_the_month = [["Month : " + str(Month)]]
        PM10_0_data_for_the_month = [["Month : " + str(Month)]]
        CO2_PPM_data_for_the_month = [["Month : " + str(Month)]]
        # Mảng lưu dữ liệu để tìm ra mode theo từng tháng ----------------------------

        Data_array_find_the_mode_of_temperature_in_the_month.append([])
        Data_array_find_the_mode_of_humidity_in_the_month.append([])
        Data_array_find_the_mode_of_pressure_in_the_month.append([])
        Data_array_find_the_mode_of_PM1_0_in_the_month.append([])
        Data_array_find_the_mode_of_PM2_5_in_the_month.append([])
        Data_array_find_the_mode_of_PM10_0_in_the_month.append([])
        Data_array_find_the_mode_of_CO2_PPM_in_the_month.append([])

        #--------------------------------------------------------------------------------------------
        max_day = day_in_month(Month,year)
        Data_array_find_the_mode_of_temperature_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_humidity_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_pressure_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_PM1_0_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_PM2_5_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_PM10_0_in_the_month[Month] = [["Month : " + str(Month)]]
        Data_array_find_the_mode_of_CO2_PPM_in_the_month[Month] = [["Month : " + str(Month)]]

        # Chạy vòng for theo ngày để đọc file có đánh dấu tên theo ngày
        # Day là vòng for chạy ngày
        for Day in range(1,max_day+1) :
            Complete_file_name = Incomplete_file_name
            # Điều chỉnh cho đúng địa chỉ file
            if Month<= 9 : 
                Complete_file_name = Complete_file_name +'0'
            Complete_file_name = Complete_file_name + str(Month) + '-'
            if Day <= 9 :
                Complete_file_name = Complete_file_name + '0'
            Complete_file_name = Complete_file_name + str(Day)  + ".txt"
            #-------------------------------------------------------------------
            Number_of_data_elements_in_the_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_temperature_data = [["Day : " + str(Day)]]
            Average_Temperature_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_humidity_data = [["Day : " + str(Day)]]
            Average_Humidity_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_pressure_data = [["Day : " + str(Day)]]
            Average_Pressure_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_PM1_0_data = [["Day : " + str(Day)]]
            Average_PM1_0_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_PM2_5_data = [["Day : " + str(Day)]]
            Average_PM2_5_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_PM10_0_data = [["Day : " + str(Day)]]
            Average_PM10_0_data__for_each_hour = [["Day : " + str(Day)]]
            #-------------------------------------------------------------------
            Hourly_CO2_PPM_data = [["Day : " + str(Day)]]
            Average_CO2_PPM_data__for_each_hour = [["Day : " + str(Day)]] 
            #-------------------------------------------------------------------
            # Kiểm tra xem file để lấy dữ liệu có tồn tại hay không bằng lệnh path.exists
            if path.exists(Complete_file_name) :
                # Test file
                print(Complete_file_name)
                # Xóa những hàng dữ liệu bị lỗi --------------------------------
                with open(Complete_file_name , 'r' , encoding= 'utf-8') as file_origianl :
                    Origianl_data_file = file_origianl.readlines()
                    Data_file_after_edit = []
                    Data_use_for_test = str(Day) + "T"
                    for data in Origianl_data_file :
                        if Data_use_for_test not in data :
                            continue 
                        else :
                            Data_file_after_edit.append(data)
                with open(Complete_file_name , 'w' , encoding= 'utf-8') as file_after_edit :
                    for data in Data_file_after_edit :
                       file_after_edit.write(data)
                
                # --------------------------------------------------------------
                # Mở file
                with open(Complete_file_name) as file :
                    Data = csv.reader(file)
                    # Tạo mảng 24 phần tử tương ứng với 24 giờ
                    for i in range(24) :
                        Hourly_temperature_data.append(0)
                        Hourly_humidity_data.append(0)
                        Hourly_pressure_data.append(0)
                        Hourly_PM10_0_data.append(0)
                        Hourly_PM1_0_data.append(0)
                        Hourly_PM2_5_data.append(0)
                        Hourly_CO2_PPM_data.append(0)
                        Number_of_data_elements_in_the_hour.append(0)
                    for row in Data :
                        #------------------------------------------------------------------------
                        try :
                            Number_of_data_elements_in_the_hour[int(row[1][11:13])+1] += 1 
                            Total_number_of_data_elements[int(row[1][11:13])] +=1
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_temperature_in_the_year[int(row[1][11:13])].append(float(row[2]))
                            Array_of_hourly_temperature_data[int(row[1][11:13])].append(float(row[2]))
                            Hourly_temperature_data[int(row[1][11:13])+1] += float(row[2])
                            Total_Temperature_data[int(row[1][11:13])] += float(row[2])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_humidity_in_the_year[int(row[1][11:13])].append(float(row[3]))
                            Array_of_hourly_humidity_data[int(row[1][11:13])].append(float(row[3]))
                            Hourly_humidity_data[int(row[1][11:13])+1] += float(row[3])
                            Total_Humidity_data[int(row[1][11:13])] += float(row[3])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_pressure_in_the_year[int(row[1][11:13])].append(float(row[4]))
                            Array_of_hourly_pressure_data[int(row[1][11:13])].append(float(row[4]))
                            Hourly_pressure_data[int(row[1][11:13])+1] += float(row[4])
                            Total_Pressure_data[int(row[1][11:13])] += float(row[4])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_PM1_0_in_the_year[int(row[1][11:13])].append(float(row[5]))
                            Array_of_hourly_PM1_0_data[int(row[1][11:13])].append(float(row[5]))
                            Hourly_PM1_0_data[int(row[1][11:13])+1] += float(row[5])
                            Total_PM1_0_data[int(row[1][11:13])] += float(row[5])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_PM2_5_in_the_year[int(row[1][11:13])].append(float(row[6]))
                            Array_of_hourly_PM2_5_data[int(row[1][11:13])].append(float(row[6]))
                            Hourly_PM2_5_data[int(row[1][11:13])+1] += float(row[6])
                            Total_PM2_5_data[int(row[1][11:13])] += float(row[6])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_PM10_0_in_the_year[int(row[1][11:13])].append(float(row[7]))
                            Array_of_hourly_PM10_0_data[int(row[1][11:13])].append(float(row[7]))
                            Hourly_PM10_0_data[int(row[1][11:13])+1] += float(row[7])
                            Total_PM10_0_data[int(row[1][11:13])] += float(row[7])
                        except :
                            continue
                        #------------------------------------------------------------------------
                        try :
                            Data_array_find_the_mode_of_CO2_PPM_in_the_year[int(row[1][11:13])].append(float(row[8]))
                            Array_of_hourly_CO2_PPM_data[int(row[1][11:13])].append(float(row[8]))
                            Hourly_CO2_PPM_data[int(row[1][11:13])+1] += float(row[8])
                            Total_CO2_PPM_data[int(row[1][11:13])] += float(row[8])
                        except :
                            continue
                        #------------------------------------------------------------------------
                    for i in range(24) :
                        # Kiểm tra mảng đếm , nếu có giá trị bằng 0 thì giá trị của các data là bằng 0
                        if Number_of_data_elements_in_the_hour[i+1] == 0 :
                            Average_Temperature_data__for_each_hour.append(0)
                            Average_Humidity_data__for_each_hour.append(0)
                            Average_Pressure_data__for_each_hour.append(0)
                            Average_PM1_0_data__for_each_hour.append(0)
                            Average_PM2_5_data__for_each_hour.append(0)
                            Average_PM10_0_data__for_each_hour.append(0)
                            Average_CO2_PPM_data__for_each_hour.append(0) 
                        else :
                        # Nếu count có giá trị khác 0 , thì thực hiện phép chia
                            Average_Temperature_data__for_each_hour.append(Hourly_temperature_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_Humidity_data__for_each_hour.append(Hourly_humidity_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_Pressure_data__for_each_hour.append(Hourly_pressure_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_PM1_0_data__for_each_hour.append(Hourly_PM1_0_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_PM2_5_data__for_each_hour.append(Hourly_PM2_5_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_PM10_0_data__for_each_hour.append(Hourly_PM10_0_data[i+1] / Number_of_data_elements_in_the_hour[i+1])
                            Average_CO2_PPM_data__for_each_hour.append(Hourly_CO2_PPM_data[i+1] / Number_of_data_elements_in_the_hour[i+1])

                # Lưu giá trị trung bình của từng giờ trong mỗi tháng 
                #-------------------------------------------------------------------------------------------
                Number_of_data_elements_in_the_month.append(Number_of_data_elements_in_the_hour)
                #-------------------------------------------------------------------------------------------
                Temperature_data_for_the_month.append(Average_Temperature_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                Humidity_data_for_the_month.append(Average_Humidity_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                Pressure_data_for_the_month.append(Average_Pressure_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                PM1_0_data_for_the_month.append(Average_PM1_0_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                PM2_5_data_for_the_month.append(Average_PM2_5_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                PM10_0_data_for_the_month.append(Average_PM10_0_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                CO2_PPM_data_for_the_month.append(Average_CO2_PPM_data__for_each_hour)
                #-------------------------------------------------------------------------------------------
                check_existence_of_data_for_the_month = True
                #-------------------------------------------------------------------------------------------
            else :
                # Thêm dữ liệu là 0 vào mảng trong 24 giờ nếu k có dữ liệu
                for j in range(24) :
                    #-------------------------------------------------------------------------------------------
                    Number_of_data_elements_in_the_hour.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_temperature_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_humidity_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_pressure_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_PM1_0_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_PM2_5_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_PM10_0_data.append(0)
                    #-------------------------------------------------------------------------------------------
                    Hourly_CO2_PPM_data.append(0)
                    #-------------------------------------------------------------------------------------------
            # Lưu giá trị là mảng 0 bởi vì k có dữ liệu
                #-------------------------------------------------------------------------------------------
                Number_of_data_elements_in_the_month.append(Number_of_data_elements_in_the_hour)
                #-------------------------------------------------------------------------------------------
                Temperature_data_for_the_month.append(Hourly_temperature_data)
                #-------------------------------------------------------------------------------------------
                Humidity_data_for_the_month.append(Hourly_humidity_data)
                #-------------------------------------------------------------------------------------------
                Pressure_data_for_the_month.append(Hourly_pressure_data)
                #-------------------------------------------------------------------------------------------
                PM1_0_data_for_the_month.append(Hourly_PM1_0_data)
                #-------------------------------------------------------------------------------------------
                PM2_5_data_for_the_month.append(Hourly_PM2_5_data)
                #-------------------------------------------------------------------------------------------
                PM10_0_data_for_the_month.append(Hourly_PM10_0_data)
                #-------------------------------------------------------------------------------------------
                CO2_PPM_data_for_the_month.append(Hourly_CO2_PPM_data)
                #-------------------------------------------------------------------------------------------

        # Lưu dữ liệu dùng để tìm mode vào mảng 12 tháng của từng thuộc tính
        
        Data_array_find_the_mode_of_temperature_in_the_month[Month].append(Array_of_hourly_temperature_data)    
        Data_array_find_the_mode_of_humidity_in_the_month[Month].append(Array_of_hourly_humidity_data)   
        Data_array_find_the_mode_of_pressure_in_the_month[Month].append(Array_of_hourly_pressure_data)   
        Data_array_find_the_mode_of_PM1_0_in_the_month[Month].append(Array_of_hourly_PM1_0_data)   
        Data_array_find_the_mode_of_PM2_5_in_the_month[Month].append(Array_of_hourly_PM2_5_data)   
        Data_array_find_the_mode_of_PM10_0_in_the_month[Month].append(Array_of_hourly_PM10_0_data)   
        Data_array_find_the_mode_of_CO2_PPM_in_the_month[Month].append(Array_of_hourly_CO2_PPM_data)   

        #Lưu dữ liệu trung bình mỗi giờ trong cả tháng ------------------------------------------------------------
        if check_existence_of_data_for_the_month == True :
            for i in range(24) :
                Average_Temperature_Data_for_month.append(Total_Temperature_data[i] / Total_number_of_data_elements[i])
                Average_Humidity_Data_for_month.append(Total_Humidity_data[i] / Total_number_of_data_elements[i])
                Average_Pressure_Data_for_month.append(Total_Pressure_data[i] / Total_number_of_data_elements[i])
                Average_PM1_0_Data_for_month.append(Total_PM1_0_data[i] / Total_number_of_data_elements[i])
                Average_PM2_5_Data_for_month.append(Total_PM2_5_data[i] / Total_number_of_data_elements[i])
                Average_PM10_0_Data_for_month.append(Total_PM10_0_data[i] / Total_number_of_data_elements[i])
                Average_CO2_PPM_Data_for_month.append(Total_CO2_PPM_data[i] / Total_number_of_data_elements[i])

            Average_Temperature_data_for_the_month.append(Average_Temperature_Data_for_month)
            Average_Humidity_data_for_the_month.append(Average_Humidity_Data_for_month)
            Average_Pressure_data_for_the_month.append(Average_Pressure_Data_for_month)
            Average_PM1_0_data_for_the_month.append(Average_PM1_0_Data_for_month)
            Average_PM2_5_data_for_the_month.append(Average_PM2_5_Data_for_month)
            Average_PM10_0_data_for_the_month.append(Average_PM10_0_Data_for_month)
            Average_CO2_PPM_data_for_the_month.append(Average_CO2_PPM_Data_for_month)
            check_existence_of_data_for_the_month = False
        else  :
            check_existence_of_data_for_the_month == False
            for i in range(24) :
                NULL.append(0)
            Average_Temperature_data_for_the_month.append(NULL)
            Average_Humidity_data_for_the_month.append(NULL)
            Average_Pressure_data_for_the_month.append(NULL)
            Average_PM1_0_data_for_the_month.append(NULL)
            Average_PM2_5_data_for_the_month.append(NULL)
            Average_PM10_0_data_for_the_month.append(NULL)
            Average_CO2_PPM_data_for_the_month.append(NULL)
        #----------------------------------------------------------------------------------------------------------

        #-------------------------------------------------------------------------------------------
        Total_number_of_data_elements_in_the_year.append(Number_of_data_elements_in_the_month)
        #-------------------------------------------------------------------------------------------
        Temperature_data_in_3D_array.append(Temperature_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        Humidity_data_in_3D_array.append(Humidity_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        Pressure_data_in_3D_array.append(Pressure_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        PM1_0_data_in_3D_array.append(PM1_0_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        PM2_5_data_in_3D_array.append(PM2_5_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        PM10_0_data_in_3D_array.append(PM10_0_data_for_the_month)
        #-------------------------------------------------------------------------------------------
        CO2_PPM_data_in_3D_array.append(CO2_PPM_data_for_the_month)
        #-------------------------------------------------------------------------------------------
    #Lưu giá trị trung bình của từng giờ trong cả 1 năm ----------------------------------------------------------
    for i in range(24) :
        if Total_number_of_data_elements[i] == 0 :
            Average_Temperature_Data_for_year.append(0)
            Average_Humidity_Data_for_year.append(0)
            Average_Pressure_Data_for_year.append(0)
            Average_PM1_0_in_year.append(0)
            Average_PM2_5_Data_for_year.append(0)
            Average_PM10_0_Data_for_year.append(0)
            Average_CO2_PPM_Data_for_year.append(0)
        else :
            Average_Temperature_Data_for_year.append(Total_Temperature_data[i] / Total_number_of_data_elements[i])
            Average_Humidity_Data_for_year.append(Total_Humidity_data[i] / Total_number_of_data_elements[i])
            Average_Pressure_Data_for_year.append(Total_Pressure_data[i] / Total_number_of_data_elements[i])
            Average_PM1_0_in_year.append(Total_PM1_0_data[i] / Total_number_of_data_elements[i])
            Average_PM2_5_Data_for_year.append(Total_PM2_5_data[i] / Total_number_of_data_elements[i])
            Average_PM10_0_Data_for_year.append(Total_PM10_0_data[i] / Total_number_of_data_elements[i])
            Average_CO2_PPM_Data_for_year.append(Total_CO2_PPM_data[i] / Total_number_of_data_elements[i])
    #-------------------------------------------------------------------------------------------------------------
```
